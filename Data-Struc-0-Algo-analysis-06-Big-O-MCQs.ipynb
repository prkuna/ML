The complexity of Fibonacci series is

    a) O(2n)
    b) O(log n)
    c) O(n2)
    d) O(n log n)

Answer: a
Explanation: Fibonacci is f(n) = f(n-1) + f(n-2), f(0) = 0, f(1) = 1. Let g(n) = 2n. Now prove inductively that f(n) > = g(n).
